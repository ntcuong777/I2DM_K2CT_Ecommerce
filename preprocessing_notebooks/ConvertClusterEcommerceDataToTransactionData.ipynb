```python
clusters = clustered_data['cluster'].unique()
n_cluster = len(clusters)
clusters
```

```python
clustered_data = pd.concat([data, clustered_data['cluster']], axis=1)
```

```python
clustered_data
```

```python
cluster_data_list = [
    clustered_data.loc[clustered_data["cluster"] == "cluster1"].drop(labels=["CustomerID", "Monetary", "Recency", "Frequency", "cluster"], axis=1),
    clustered_data.loc[clustered_data["cluster"] == "cluster2"].drop(labels=["CustomerID", "Monetary", "Recency", "Frequency", "cluster"], axis=1),
    clustered_data.loc[clustered_data["cluster"] == "cluster3"].drop(labels=["CustomerID", "Monetary", "Recency", "Frequency", "cluster"], axis=1)
]
```

```python
cluster_data_list
```

```python
cluster_data_list = [i.applymap(lambda x: 1 if x > 0 else 0) for i in cluster_data_list]
```

```python
cluster_data_list
```

```python
for i in range(len(cluster_data_list)):
    cluster_data_list[i].to_csv("/content/cluster_" + str(i+1) + "_trans_data.csv", index=False)
```
